```python
import pandas as pd

normalized_df = df.copy()
```

```python
normalized_df["remote"] = (
    normalized_df["remote"]
    .str.lower()
    .map({"yes": True, "no": False})
)

normalized_df["adaptive"] = (
    normalized_df["adaptive"]
    .str.lower()
    .map({"yes": True, "no": False})
)
```

```python
normalized_df[["remote","adaptive"]].head()
```

```python
import re

def extract_minutes(text):

    match = re.search(r"\d+", str(text))

    if match:
        return int(match.group())

    return None

normalized_df["duration_minutes"] = normalized_df["duration"].apply(extract_minutes)
```

```python
normalized_df[
["duration","duration_minutes"]
].head()
```

```python
list_columns = [
    "job_levels",
    "languages",
    "keys"
]

for col in list_columns:

    normalized_df[col] = normalized_df[col].apply(
        lambda x: x if isinstance(x,list) else []
    )
```

```python
columns_to_remove = [

    "scraped_at",

    "duration_raw",

    "languages_raw",

    "job_levels_raw",

    "status"

]

normalized_df = normalized_df.drop(
    columns=columns_to_remove
)
```

```python
def build_search_text(row):

    return f"""
Assessment Name:
{row['name']}

Description:
{row['description']}

Categories:
{', '.join(row['keys'])}

Job Levels:
{', '.join(row['job_levels'])}

Languages:
{', '.join(row['languages'])}

Duration:
{row['duration']}

Remote:
{row['remote']}

Adaptive:
{row['adaptive']}
""".strip()
```

```python
normalized_df["search_text"] = normalized_df.apply(
    build_search_text,
    axis=1
)
```

```python
print(
normalized_df.iloc[0]["search_text"]
)
```

```python
def build_metadata(row):

    return {

        "entity_id": row["entity_id"],

        "name": row["name"],

        "url": row["link"],

        "duration": row["duration_minutes"],

        "remote": row["remote"],

        "adaptive": row["adaptive"],

        "job_levels": row["job_levels"],

        "languages": row["languages"],

        "categories": row["keys"]

    }
```

```python
normalized_df["metadata"] = normalized_df.apply(
    build_metadata,
    axis=1
)
```

```python
from collections import Counter

# Category distribution
category_counter = Counter()
for categories in normalized_df["keys"]:
    category_counter.update(categories)

# Job level distribution
job_counter = Counter()
for levels in normalized_df["job_levels"]:
    job_counter.update(levels)

# Language distribution
language_counter = Counter()
for langs in normalized_df["languages"]:
    language_counter.update(langs)

profile = {
    "total_assessments": len(normalized_df),
    "assessment_categories": dict(category_counter),
    "job_levels": dict(job_counter),
    "languages": dict(language_counter),
    "average_duration": normalized_df["duration_minutes"].mean(),
    "minimum_duration": normalized_df["duration_minutes"].min(),
    "maximum_duration": normalized_df["duration_minutes"].max()
}

profile
```
